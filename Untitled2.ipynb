```python
# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------------
# Sample Food Order Dataset
# ---------------------------------
data = {
    'Order_ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Time': ['10:00', '12:30', '13:00', '19:00', '20:00', '12:00', '18:30', '20:30'],
    'Location': [
        'Chennai', 'Madurai', 'Chennai', 'Coimbatore',
        'Chennai', 'Madurai', 'Coimbatore', 'Chennai'
    ],
    'Item': [
        'Pizza', 'Burger', 'Pizza', 'Biryani',
        'Pizza', 'Burger', 'Biryani', 'Pizza'
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# ---------------------------------
# Display Dataset
# ---------------------------------
print("===== ORDER DATA =====")
print(df)

# ---------------------------------
# 1. Peak Order Times Analysis
# ---------------------------------
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour

peak_orders = df.groupby('Hour').size()

print("\n===== PEAK ORDER TIMES =====")
print(peak_orders)

# ---------------------------------
# 2. Most Popular Dishes Analysis
# ---------------------------------
popular_dishes = df['Item'].value_counts()

print("\n===== MOST POPULAR DISHES =====")
print(popular_dishes)

# ---------------------------------
# 3. Orders by Location Analysis
# ---------------------------------
location_orders = df['Location'].value_counts()

print("\n===== ORDERS BY LOCATION =====")
print(location_orders)

# ---------------------------------
# 4. Dashboard Visualization
# ---------------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Chart 1: Peak Order Times
peak_orders.plot(
    kind='bar',
    color='skyblue',
    ax=axes[0]
)
axes[0].set_title('Peak Order Times')
axes[0].set_xlabel('Hour')
axes[0].set_ylabel('Number of Orders')

# Chart 2: Most Popular Dishes
popular_dishes.plot(
    kind='bar',
    color='orange',
    ax=axes[1]
)
axes[1].set_title('Most Popular Dishes')
axes[1].set_xlabel('Food Item')
axes[1].set_ylabel('Orders')

# Chart 3: Orders by Location
location_orders.plot(
    kind='bar',
    color='green',
    ax=axes[2]
)
axes[2].set_title('Orders by Location')
axes[2].set_xlabel('Location')
axes[2].set_ylabel('Orders')

plt.tight_layout()
plt.show()
```
